```python
import numpy as np
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt 

# Cargar los datos de California Housing
all_data = fetch_california_housing()
x_data = all_data.data
y_data = all_data.target


# Tamao de los puntos en los diagramas de dispersin
point_size = 10

# Diagrama de dispersin MedInc vs VALUE (variable objetivo)
plt.figure(figsize=(8, 4))
plt.scatter(x_data[:, 0], y_data, alpha=0.5, color='b', label='MedInc vs VALUE', s=point_size)
plt.xlabel('MedInc')
plt.ylabel('VALUE')
plt.title('Diagrama de dispersin MedInc vs VALUE')
plt.legend()
plt.grid(True)
plt.savefig('MedInc_vs_VALUE.png', dpi=300)
plt.show()

# Diagrama de dispersin Population vs VALUE (variable objetivo)
plt.figure(figsize=(8, 4))
plt.scatter(x_data[:, 4], y_data, alpha=0.5, color='b', label='Population vs VALUE', s=point_size)
plt.xlabel('Population')
plt.ylabel('VALUE')
plt.title('Diagrama de dispersin Population vs VALUE')
plt.legend()
plt.grid(True)
plt.savefig('Population_vs_VALUE.png', dpi=300)
plt.show()

# Diagrama de dispersin HouseAge vs VALUE (variable objetivo)
plt.figure(figsize=(8, 4))
plt.scatter(x_data[:, 1], y_data, alpha=0.5, color='b', label='HouseAge vs VALUE', s=point_size)
plt.xlabel('HouseAge')
plt.ylabel('VALUE')
plt.title('Diagrama de dispersin HouseAge vs VALUE')
plt.legend()
plt.grid(True)
plt.savefig('HouseAge_vs_VALUE.png', dpi=300)
plt.show()

# Diagrama de dispersin AveRooms vs VALUE (variable objetivo)
plt.figure(figsize=(10, 4))
plt.scatter(x_data[:, 2], y_data, alpha=0.5, color='b', label='AveRooms vs VALUE', s=point_size)
plt.xlabel('AveRooms')
plt.ylabel('VALUE')
plt.title('Diagrama de dispersin AveRooms vs VALUE')
plt.legend()
plt.grid(True)
plt.savefig('AveRooms_vs_VALUE.png', dpi=300)
plt.show()

```
