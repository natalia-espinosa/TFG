```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset de California Housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names) 
df['Price'] = housing.target

# Calcular la matriz de correlación
correlation_matrix = df.corr()

# Mostrar la matriz de correlación
plt.figure(figsize=(10, 8))  # Aumentar el tamaño de la figura
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10}, cbar_kws={'shrink': .8})  # Ajustar el tamaño de las anotaciones y la barra de color
plt.xticks(rotation=45, ha='right')  # Rotar los nombres de las columnas
plt.yticks(rotation=0)  # Asegurar que los nombres de las filas no estén rotados
plt.title('Matriz de correlación del dataset California Housing', fontsize=16)
plt.savefig('MatrizCorrelacion.png', dpi=300)
plt.show()
```
