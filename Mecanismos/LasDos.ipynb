```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, norm

# Parámetros para las distribuciones
epsilon = 1
delta = 1e-5
sensitivity = 1

# Generación de datos para la distribución Laplace
x = np.linspace(-10, 10, 1000)  # Valores de x
laplace_noise = laplace.pdf(x, scale=sensitivity / epsilon)  # Densidad de probabilidad para Laplace

# Generación de datos para la distribución Gaussiana
sigma = np.sqrt(2 * np.log(1.25 / delta)) * sensitivity / epsilon  # Calculo de sigma para la distribución Gaussiana
gaussian_noise = norm.pdf(x, scale=sigma)  # Densidad de probabilidad para Gaussiana

# Creación del gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, laplace_noise, label='Laplace', color='r')
plt.plot(x, gaussian_noise, label='Gaussian', color='b')
plt.xlabel('e')
plt.ylabel('Densidad de Probabilidad')
plt.title(' Distribuciones Laplace vs Gaussiana')
plt.legend()
plt.grid(True)
plt.show()
```
