```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución Gaussiana
mu = 0      # Media
sigma = 1   # Desviación estándar

# Generar valores de x
x = np.linspace(-5, 5, 1000)

# Calcular la función de densidad de probabilidad para la distribución Gaussiana
gaussian = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, gaussian, label=f'Gaussian Distribution\n($\mu$ = {mu}, $\sigma$ = {sigma})', color='r')
plt.title('Distribución Gaussiana')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)

# Guardar el gráfico en un archivo PNG
plt.savefig('gaussian_distribution.png', dpi=300)

# Mostrar el gráfico en pantalla
plt.show()
```
