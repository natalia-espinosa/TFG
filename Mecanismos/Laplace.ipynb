```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetro de escala
b = 0.5

# Generar valores de x
x = np.linspace(-5 , 5 , 1000)

laplace = (1 / (2 * b)) * np.exp(-np.abs(x) / b)

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, laplace, label=f'Laplace Distribution\n(b = {b})', color= 'r')
plt.title('Distribución de Laplace')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)

# Guardar el gráfico en un archivo PNG
plt.savefig('laplace_distribution.png', dpi=300)

# Mostrar el gráfico en pantalla
plt.show()
```
